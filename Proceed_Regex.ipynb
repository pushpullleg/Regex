# Regex Metacharacters

```python
re.search("[0-9][0-9][0-9]",Strings)
```

```python
re.search(r"[0-9]",Strings)
```

### Identify pattern in a string

```python
re.search(r"\d\d\d",Strings)
```

```python
re.search(r"\D","78u8")
```

### Meta characters. 

```python
re.search("cs..","cs5454545")
```

```python
re.search("C..",Strings)
```

```python
re.search("c.ci543","csci543csci333")

```

```python
re.findall("c.ci543","csci543,csci333")
```

```python
re.sub("CSCI432","csci543",Strings)
```

```python
My_Name = "Mukesh"

if re.search(My_Name, Strings):
    print("Found")
    Capital_Name=str.capitalize(My_Name)
    re.sub(My_Name,Capital_Name,Strings)
else:
    print("Not Found")

print(Strings[0:6])
```

```python
My_Name = "Mukesh"
TARGET = "MUKESH"  # Desired fully capitalized form

# Replace every occurrence of the name in any case with fully uppercase
if re.search(My_Name, Strings, flags=re.IGNORECASE):
    Strings = re.sub(r"mukesh", TARGET, Strings, flags=re.IGNORECASE)
    print("Replaced all occurrences with", TARGET)
else:
    print("Name not found.")

print(Strings)
```
